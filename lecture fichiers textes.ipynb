```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from math import sqrt,pi,exp
```

```python
df = pd.read_csv("2018-11 JMT Li-rich NMC Pristine-csp-22112018 (1).TXT", sep = "\t", skiprows = 3)
```

```python
df
```

```python
BE=[]
BE=df["BE_LiM3p_s MRSof/23"]
BE
```

```python
def gaussienne(x,mu,sigma_carre):
    return (1/sqrt(2*pi*sigma_carre)*exp(-1/2*(x-mu)**2/sigma_carre))
```

```python
G=[]
for i in range (len(BE)):
    y=gaussienne(BE[i], 64, 11)
    G.append(y)
```

```python
G
```

```python
plt.plot(BE, G)
```
